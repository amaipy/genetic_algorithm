#### Importing necessary depedencies

```python
from tp1 import evolution, gen_plot
import pandas as pd
```

#### Setting up params

```python
gen_size=100
max_depth=7
min_depth=3
cross_rate=0.6
mut_rate=0.3
tour_size=3
print_stats = False
```

#### Selecting the circle dataset

```python
input_path = "./datasets-TP1/SR_circle.txt"
```

#### Runing first tests, varying the population sizes

```python
pop_size = [25, 50, 100]

all_best_f = []
all_duplicates = []

for pop in pop_size:
    best_f, duplicates = evolution(input_path, gen_size, pop, max_depth, min_depth, cross_rate, mut_rate, tour_size, print_stats)[2:4]
    all_best_f.append(best_f)
    all_duplicates.append(duplicates)
```

#### Organizing results for plotting

```python
gen_list = list(range(gen_size))
list_df = []

for i in range(len(pop_size)):
    df = pd.DataFrame([gen_list, all_best_f[i], all_duplicates[i]], index=['Generation', 'Fitness', 'Duplicates']).T
    df['Type'] = f'Pop # = {pop_size[i]}'
    list_df.append(df)

list_df = pd.concat(list_df)
```

#### Best fitness values for each Population values (25, 50 or 100 individuals) and 100 Generations

```python
gen_plot(list_df, 'Fitness')
```

#### # of duplicates for each Population values (25, 50 or 100 individuals) and 100 Generations

```python
gen_plot(list_df, 'Duplicates')
```

#### Fixing population size and setting a new generation size based on the previous results (pop = 50, gen = 20)
#### varying crossover and mutation rates

```python
pop_size = 50
gen_size = 20

cross_rate = [.9, .6, .5]
mut_rate = [.1, .3, .4]
```

```python
all_best_f = []
all_duplicates = []
all_best_xover = []

for i in range(len(cross_rate)):
    best_f, duplicates, _, best_xover = evolution(input_path, gen_size, pop_size, max_depth, min_depth, cross_rate[i], mut_rate[i], tour_size, print_stats)[2:6]
    all_best_f.append(best_f)
    all_duplicates.append(duplicates)
    all_best_xover.append(best_xover)
```

```python
gen_list = list(range(gen_size))
list_df = []

for i in range(len(cross_rate)):
    df = pd.DataFrame([gen_list, all_best_f[i], all_duplicates[i], all_best_xover[i]], index=['Generation', 'Fitness', 'Duplicates', 'Best Children Xover']).T
    df['Type'] = f'Mutation % = {mut_rate[i]*100}; Xover % = {cross_rate[i]*100}'
    list_df.append(df)

list_df = pd.concat(list_df)
```

#### Best fitness values for each xover and mutation rates

```python
gen_plot(list_df, 'Fitness')
```

#### # of duplicates for each xover and mutation rates

```python
gen_plot(list_df, 'Duplicates')
```

#### # of best children from xover

```python
gen_plot(list_df, 'Best Children Xover')
```

#### Fixing all params based on previous results (tour = 7)
#### comparing noise and normal dataset

```python
tour_size = 7
input_path = ["./datasets-TP1/SR_circle.txt", "./datasets-TP1/SR_ellipse_noise.txt"]

all_best_f = []

for path in input_path:
    best_f = evolution(path, gen_size, pop, max_depth, min_depth, cross_rate, mut_rate, tour_size, print_stats)[2]
    all_best_f.append(best_f)
```

```python
gen_list = list(range(gen_size))
list_df = []

for i in range(len(input_path)):
    df = pd.DataFrame([gen_list, all_best_f[i]], index=['Generation', 'Fitness']).T
    input_str = input_path[i].replace("./datasets-TP1/", '').replace(".txt", '')
    df['Type'] = f'Input = {input_str}'
    list_df.append(df)

list_df = pd.concat(list_df)
```

#### Best fitness values for normal and noise dataset

```python
gen_plot(list_df, 'Fitness')
```
